```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Oguntuga\Downloads\Africa_climate_change.csv')
```

```python
# For demonstration, create a mock dataset
date_range = pd.date_range(start='1980-01-01', end='2023-12-31', freq='D')
```

```python
# Create data for each country with the same number of entries
data = {
    'Date': date_range,
    'Tunisia': (20 + 10 * np.sin(np.linspace(0, 3 * np.pi, len(date_range)))).tolist(),
    'Cameroon': (25 + 10 * np.sin(np.linspace(0, 3 * np.pi, len(date_range)))).tolist(),
    'Senegal': (27 + 5 * np.sin(np.linspace(0, 3 * np.pi, len(date_range)))).tolist(),
    'Egypt': (30 + 7 * np.sin(np.linspace(0, 3 * np.pi, len(date_range)))).tolist(),
    'Angola': (28 + 8 * np.sin(np.linspace(0, 3 * np.pi, len(date_range)))).tolist(),
}


```

```python
# Create DataFrame from the mock data
df = pd.DataFrame(data)
df.set_index('Date', inplace=True)
```

```python
# Plot line chart for Tunisia and Cameroon
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Tunisia'], label='Tunisia', color='blue')
plt.plot(df.index, df['Cameroon'], label='Cameroon', color='orange')
plt.title('Average Temperature Fluctuations (1980 - 2023)')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Zoom in on the specific time frame (1980 to 2005)
plt.figure(figsize=(12, 6))
df_zoomed = df['1980':'2005']
plt.plot(df_zoomed.index, df_zoomed['Tunisia'], label='Tunisia', color='blue')
plt.plot(df_zoomed.index, df_zoomed['Cameroon'], label='Cameroon', color='orange')
plt.title('Average Temperature Fluctuations (1980 - 2005)')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
# Create histograms for Senegal between [1980, 2000] and [2000, 2023]
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Senegal']['1980':'2000'], bins=20, kde=True)
plt.title('Temperature Distribution in Senegal (1980 - 2000)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['Senegal']['2000':'2023'], bins=20, kde=True)
plt.title('Temperature Distribution in Senegal (2000 - 2023)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Select the best chart to show Average temperature per country
# Bar plot of average temperatures
average_temps = df.mean()
plt.figure(figsize=(10, 6))
average_temps.plot(kind='bar', color='skyblue')
plt.title('Average Temperature per Country (1980 - 2023)')
plt.xlabel('Country')
plt.ylabel('Average Temperature (°C)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
